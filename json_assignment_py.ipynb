```python
## Task 1 — Build a JSON Structure

#Step 1: Importing json and Create a Python dictionary
import json

user_profile = {
    "name": "John",
    "age": 28,
    "email": "john@example.com",
    "is_active": True,
    "skills": ["Python", "SQL", "Machine Learning"]
}
```

```python
#Step 2: Convert dictionary to JSON string
json_data = json.dumps(user_profile, indent=4)

#Step 3: Print Json
print(json_data)
```

```python
#Task 2 — Parse an API Response

#Step 1: Json string from API

import json

api_response = '{"status": "success", "data": {"user_id": 101, "username": "alex99", "score": 87.5}}'
```

```python
#Step 2: Convert JSON string → Python dictionary
response = json.loads(api_response)

#Step 3: Extract values
username = response["data"]["username"]
score = response["data"]["score"]

#Step 4: Print
print("Username:", username)
print("Score:", score)
print(f"User {username} scored {score} points")
```
